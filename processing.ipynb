## Trimming down the data

After scraping through all the tweets, they can now be manipulated to remove any dates and times that aren't relevant.

The easiest way to start this is by writing all the data in the csv to a variable that can be manipulated.

The function `read_to_variable` takes the name of the user and the start year that tweets were scraped from, and returns a list of all the tweet data.

Note: The files are stored with the name format: *name*-after*year*.csv. If you are looking for the file that has all of Elon Musk's tweets after 2019, the calling function would be:
`read_to_variable("elonmusk",2019)`.

```python
NAME = "elonmusk"
YEAR = 2019

tweets = read_to_variable(NAME,YEAR)

print(tweets)
```

With all the data in rows, it can then be iterated through to only include tweets that meet specific criteria. That way there is a much smaller dataset to work with. For example, if you only want tweets tweeted on a certain date to be in the list, this can be done.

The function `show_tweets_on` takes a list of tweets and a target date as the input, and returns the tweets that were tweeted on that date.

Note: The date should be in the format `mm-dd-yyyy`

```python
NAME = "elonmusk"
YEAR = 2019
DATE = "06-11-2022"
tweets = read_to_variable(NAME,YEAR)
specific_tweets = show_tweets_on(tweets,DATE)
print(specific_tweets)
```
